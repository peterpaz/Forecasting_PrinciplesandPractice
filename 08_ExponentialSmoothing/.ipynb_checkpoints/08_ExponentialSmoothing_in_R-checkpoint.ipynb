# 1. Simple Exponential Smoothing

- This method is suitable for forecasting data with no clear trend or seasonal pattern.

```r
global_economy
```

```r
algeria_economy <- global_economy %>%
 filter(Country == "Algeria")

algeria_economy
```

```r
algeria_economy %>%
 autoplot(Exports)+
  labs(y = "% of GDP", title = "Exports: Algeria")
```

- This graph does not display any clear trending behaviour or any seasonality.

Initially the possible methos for forecasting are:
- Naive method.
- Average method.
Both can be seen as extreme weighted average.

Alternatively:
- Simple exponential smoothing: Forecasts are calculated using weighted averages, where the weights decrease exponentially as observations come from further in the past — the smallest weights are associated with the oldest observations.
- This method is an weighted average form.
- This method can be written as component form: forecast equation and smoothing equation.




```r
fit <- algeria_economy %>%
 model(
      ETS( Exports ~ error("A")+trend("N")+season("N"))
      ) 
fit
```

```r
report(fit)
```

```r
fc <- fit %>%
 forecast(h=5)

fc
```

```r
augment(fit)
```

```r
fc %>%
  autoplot(algeria_economy) +
  geom_line(aes(y = .fitted), col="#D55E00",
            data = augment(fit))
 labs(y="% of GDP", title="Exports: Algeria") +
  guides(colour = FALSE)
```

# 2. Methods with trend

## Holt’s linear trend method or Double Exponential Smoothing


```r
aus_economy <- global_economy %>%
 filter(Code=="AUS") %>%
 mutate(Pop= Population/1e6)
```

```r
aus_economy
```

```r
autoplot(aus_economy,Pop)+
labs(y = "Millions", title = "Australian population")
```

```r
fit <-aus_economy %>%
 model(
      AAN=ETS(Pop ~ error("A")+trend("A")+season("N"))
      )
```

```r
fit
```

```r
report(fit)
```

```r
augment(fit)
```

```r
fc <- fit %>% forecast(h=10)
```

```r
fc
```

```r
fc %>%
  autoplot(aus_economy) +
  geom_line(aes(y = .fitted), col="#D55E00",
            data = augment(fit))+
 labs(y="Population", title="Australia") +
  guides(colour = FALSE)
```

## Damped trend methods- Holt’s linear trend method WITH DAMPED


```r
fit <-aus_economy %>%
 model(
  `Holt's method`        = ETS(Pop ~ error("A")+ trend("A")+ season("N") ),
  `Damped Holt's method` = ETS(Pop ~ error("A")+ trend("Ad", phi=0.9)+ season("N") ),
      )
fit
```

```r
report(fit)
```

```r
augment(fit)
```

```r
fc <- fit %>% forecast(h=15)
fc
```

```r
fit %>% forecast(h=15) %>%
autoplot(aus_economy, level = NULL) +
  labs(title = "Australian population",
       y = "Millions") +
  guides(colour = guide_legend(title = "Forecast"))
```

# Time series cross-validation to compare the one-step forecast accuracy

```r
www_usage %>%
  stretch_tsibble(.init = 10)
```

```r
fit <- www_usage %>%
  stretch_tsibble(.init = 10) %>%
  model(
    SES = ETS(value ~ error("A") + trend("N") + season("N")),
    Holt = ETS(value ~ error("A") + trend("A") + season("N")),
    Damped = ETS(value ~ error("A") + trend("Ad") +
                   season("N"))
  )
```

```r
report(fit)
```

```r
augment(fit)
```

```r
fit %>%
  forecast(h = 1) %>%
  accuracy(www_usage)
```

- Damped Holt’s method is best whether you compare MAE or RMSE values.
- Given the Damped Holt’s method, we use whole data to get forecast

```r
fit <- www_usage %>%
  model(
    Damped = ETS(value ~ error("A") + trend("Ad") +
                   season("N"))
  )
```

```r
report(fit)
```

```r
tidy(fit)
```

```r
fit %>%
 forecast(h=10) %>%
 autoplot(www_usage)+
   labs(x="Minute", y="Number of users",
       title = "Internet usage per minute")
```

# Holt-Winters’ damped method
## Example: Holt-Winters method with daily data

- Holt-Winters method can also be used for daily type of data
- where the seasonal period is $m=7$  , and the appropriate unit of time for $h$ is in days.

This data is about pedestrian traffic at a busy Melbourne train station in July 2016.

```r
pedestrian
```

```r
sth_cross_ped <- pedestrian %>%
 filter(Date >= "2016-07-01", Sensor == "Southern Cross Station") %>%
 index_by(Date)
```

```r
sth_cross_ped
```

```r
sth_cross_ped %>% summarise(Count =sum(Count)/1000)
```

```r
sth_cross_ped %>%
  filter(Date <= "2016-07-31") %>%
  model(
  hw = ETS(Count ~ error("M")+trend("Ad")+season("M"))
       ) 
```

```r
sth_cross_ped <- pedestrian %>%
  filter(Date >= "2016-07-01",
         Sensor == "Southern Cross Station") %>%
  index_by(Date) %>%
  summarise(Count = sum(Count)/1000)

sth_cross_ped %>%
  filter(Date <= "2016-07-31") %>% # TRAINING DATA
  model(
      hw = ETS(Count ~ error("M")+trend("Ad")+season("M"))
       ) %>%
 forecast(h = "2 weeks")%>%
  autoplot(sth_cross_ped %>% filter(Date <= "2016-08-14")) + # TEST DATA
 labs(title = "Daily traffic: Southern Cross",
       y="Pedestrians ('000)")
```

- Clearly the model has identified the weekly seasonal pattern and the increasing trend at the end of the data, and the forecasts are a close match to the test data.

# 4. Estimation and Model Selection

## Model selection
- information criteria can be used for model selection.
- The AIC, AICc and BIC can be used to determine which of the ETS models is most appropriate for a given time series.

## Example: Domestic holiday tourist visitor nights in Australia
- data: Australian holiday tourism over the period 2016–2019

BEFORE:



```r
aus_holidays <- tourism %>%
  filter(Purpose == "Holiday") %>%
  summarise(Trips = sum(Trips)/1e3)
fit <- aus_holidays %>%
  model(
    additive = ETS(Trips ~ error("A") + trend("A") +
                                                season("A")),
    multiplicative = ETS(Trips ~ error("M") + trend("A") +
                                                season("M"))
  )

fit %>% select(additive) %>% report(fit)
```

```r
fit %>% select(multiplicative) %>% report(fit)
```

NOW: AUTOMATIC SELECTION OF THE MODEL

```r
aus_holidays <- tourism %>%
  filter(Purpose == "Holiday") %>%
  summarise(Trips = sum(Trips)/1e3)
fit <- aus_holidays %>%
  model(ETS(Trips))
report(fit)
```

```r
components(fit) 
```

```r
components(fit) %>%
autoplot()+
labs(title = "ETS(M,N,A) components")
```

- Because this model has multiplicative errors, the innovation residuals are not equivalent to the regular residuals (i.e., the one-step training errors).
- The innovation residuals are given by  $\hat{\epsilon_t}$, while the regular residuals are defined as  $y_t -\hat{y}_{t|t-1}$

 

```r
augment(fit)
```

```r
augment(fit)%>%
autoplot(.resid	)
```

```r
augment(fit)%>%
autoplot(.innov)
```
